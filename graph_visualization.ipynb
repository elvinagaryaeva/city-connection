```python
import networkx as nx
import matplotlib.pyplot as plt
# import seaborn as sns
%matplotlib inline
# sns.set()

G=nx.Graph()

# for i in range(0,9):
#     G.add_node(i)

for i in range(0,9):
    G.add_node(dd[i])

# G.add_nodes_from(range(0,9))

for i in range(0,8):
    for j in range(i+1,9):
        if a[i][j] == '1':
            G.add_edge(dd[i],dd[j])
            

# weight = nx.get_edge_attributes(G,'weight')
# pos = nx.get_edge_attributes(G,'pos')
```

```python
nx.draw_networkx(G,with_labels=True)
plt.draw()
plt.show()
```
